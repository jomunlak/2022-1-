```python
import numpy as np
import pandas as pd
import mglearn
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
iris = load_iris()
```

```python
X = iris.data[:,:2]
y = iris.target

model = LogisticRegression()
model.fit(X,y)
y_pred = model.predict(X)
y_pred
mglearn.plots.plot_2d_classification(model, X, fill=True, eps = 0.5, alpha=0.4)
mglearn.discrete_scatter(X[:, 0], X[:,1],y)
plt.legend(loc=4)
```

```python
# train 데이터와 test 데이터를 자동으로 나누어주는 train_test_split() 메서드
X_train, X_test, y_train, y_test = train_test_split(iris.data[:,:2], iris.target)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 두 데이터를 산점도로 표현해보자
plt.subplot(1,2,1)
plt.scatter(X_train[:,0], X_train[:,1], c=y_train)
plt.subplot(1,2,2)
plt.scatter(X_test[:,0], X_test[:,1], c=y_test)

# 좌, train 시킬 데이터       우, test 할 데이터
```

```python
model = LogisticRegression()
model.fit(X_train, y_train) # 학습할 데이터로만 학습시킴
y_pred = model.predict(X_test) 


plt.legend(loc=4)
```

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


mglearn.plots.plot_2d_classification(model, X_test, fill=True, eps = 0.5, alpha=0.4)
mglearn.discrete_scatter(X_test[:, 0], X_test[:,1],y_test) 

model.score(X_test, y_test)
```

```python
model = KNeighborsClassifier(10)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


mglearn.plots.plot_2d_classification(model, X_test, fill=True, eps = 0.5, alpha=0.4)
mglearn.discrete_scatter(X_test[:, 0], X_test[:,1],y_test) 

model.score(X_test, y_test)
```
